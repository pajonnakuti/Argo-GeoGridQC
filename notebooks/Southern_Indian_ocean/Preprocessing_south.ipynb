```python
import pandas as pd

grid_id = 113

file_path = rf"C:\INCOiS\ARGO\DATA\South_Data\Grid_Parquet\grid_{grid_id}.parquet"

df = pd.read_parquet(file_path)

print(df.shape)
print(df.head())
```

```python
df.columns
```

```python
df["TEMP_QC"] = (
    df["TEMP_QC"]
    .astype(str)
    .str.replace("b'", "", regex=False)
    .str.replace("'", "", regex=False)
)

df["PSAL_QC"] = (
    df["PSAL_QC"]
    .astype(str)
    .str.replace("b'", "", regex=False)
    .str.replace("'", "", regex=False)
)
df["DATE"] = pd.to_datetime(
    df["DATE"].astype(str),
    format="%Y%m%d%H%M%S",
    errors="coerce"
)

print(df[["DATE"]].head())
print(df.head(15))
```

```python
from seawater import dpth

df["DEPTH"] = dpth(
    df["PRES"].values,
    df["LAT"].values
)
print(df)
```

```python
import numpy as np
import pandas as pd

bins = [0, 50,200, 500, 1000, 2000, 6000]
labels = ["0-50", "50-200", "200-500", "500-1000", "1000-2000", "2000-6000"]

df["DEPTH_BIN"] = pd.cut(df["DEPTH"], bins=bins, labels=labels, include_lowest=True)
```

```python
df["TEMP_Z"] = df.groupby("DEPTH_BIN")["TEMP"].transform(
    lambda x: (x - x.mean()) / x.std()
)
```

```python
df["PSAL_Z"] = df.groupby("DEPTH_BIN")["PSAL"].transform(
    lambda x: (x - x.mean()) / x.std()
)
```

```python

df["YEAR"] = df["DATE"].dt.year
df["MONTH"] = df["DATE"].dt.month
df["DAY"] = df["DATE"].dt.day
df["SEASON"] = (df["MONTH"] % 12 + 3) // 3
df.head()
```

```python
season_map = {
    1: "Winter",
    2: "Spring",
    3: "Summer",
    4: "Autumn"
}

df["SEASON_NAME"] = df["SEASON"].map(season_map)
df.head()
```

```python
def apply_zscore_flags(df):
    """
    Convert Z-scores into QC flags.
    """
    print("\n--- STEP 7: APPLYING Z-SCORE FLAGS ---")

    def zscore_to_flag(z):
        if pd.isna(z):
            return 9
        abs_z = abs(z)
        if abs_z <= 2.0:
            return 1
        elif abs_z <= 3.0:
            return 2
        elif abs_z <= 4.0:
            return 3
        else:
            return 4

    # Apply flags
    print("Applying temperature Z-score flags...")
    df['TEMP_Z_FLAG'] = df['TEMP_Z'].apply(zscore_to_flag)

    print("Applying salinity Z-score flags...")
    df['PSAL_Z_FLAG'] = df['PSAL_Z'].apply(zscore_to_flag)

    # Final flag = worst of temp and sal
    df['Z_FINAL_FLAG'] = df[['TEMP_Z_FLAG', 'PSAL_Z_FLAG']].max(axis=1)

    # Add human readable label
    flag_labels = {
        1: 'GOOD',
        2: 'PROBABLY_GOOD',
        3: 'PROBABLY_BAD',
        4: 'BAD',
        9: 'MISSING'
    }
    df['Z_FLAG_LABEL'] = df['Z_FINAL_FLAG'].map(flag_labels)

    # Print distributions

    print(f"\nFinal Z-flag distribution:")
    for flag, count in df['Z_FINAL_FLAG'].value_counts().sort_index().items():
        label = flag_labels[flag]
        pct = 100 * count / len(df)
        print(f" Flag {flag} ({label:15s}): {count:>12,} ({pct:.2f}%)")

    print("\n✅ Z-score flags applied")

    return df
df = apply_zscore_flags(df)
```
